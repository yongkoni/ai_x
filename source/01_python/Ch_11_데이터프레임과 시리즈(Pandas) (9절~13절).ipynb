```python
# 연습하기 (apply, map)
# new_df의 연도월 데이터를 연도와 월을 int로 분리하여 컬럼을 추가하세요
# ★ 1차원 시리즈로 처리

new_df.sample()
```

```python
new_df['연도'] = new_df['연도월'].apply(lambda year : int(year.split('년')[0]))
new_df['월'] = new_df['연도월'].map(lambda month : int(month.split('년')[1][:-1]))
new_df.info()
```

```python
new_df.sample()
```

```python
new_df.drop(['연도','월'], axis=1, inplace = True)
new_df.sample()
```

```python
# ★ 2차원 데이터 프레임의 apply 이용하는 방법
```

```python
def year_month_append(df):
    df_copy = df.copy()  # 깊은 복사
    df_copy['연도'] = int(df_copy['연도월'].split('년')[0])
    df_copy['월'] = int(df_copy['연도월'].split('년')[1].replace('월', ''))
    return df_copy
```

```python
year_month_append(new_df.iloc[0])
```

```python
new_df = new_df.apply(year_month_append, axis = 1)
new_df.sample()
```

# 12절. 일괄변경하기 (결측치, 특정값)

## 12.1 fillna

- 대부분 결측치는 drop하거나, 평균값, 중위수, 그룹별 평균값, 그룹별 중위수, 이전값 등으로 대체

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3,      4, np.nan, 1],
                   [np.nan, 3, np.nan, 5],
                   [np.nan, 3, np.nan, 4]],
                 columns = [list('ABCD')])
df
```

```python
# 모든 결측치를 0으로 대체

df.fillna(0)
```

```python
# method = 'ffill' → 결측치를 이전행의 값으로 대체

df.fillna(method = 'ffill')
```

```python
# method = 'bfill' → 결측치를 다음행의 값으로 대체

df.fillna(method = 'bfill')
```

```python
# 결측치를 열마다 특정값으로 대체

values = {'A':85, 'B':80, 'C':75, 'D':90}
df.fillna(value = values)
```

```python
# 결측치를 열별 평균값으로 대체

df.fillna(value = df.mean(axis = 0))
```

```python
# 2번째 결측치까지만 결측치를 대체

df.fillna(value = values, limit = 2)
```
