## 12.1 fillna

- 대부분 결측치는 drop하거나, 평균값, 중위수, 그룹별 평균값, 그룹별 중위수, 이전값 등으로 대체

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3,      4, np.nan, 1],
                   [np.nan, 3, np.nan, 5],
                   [np.nan, 3, np.nan, 4]],
                 columns = [list('ABCD')])
df
```

```python
# 모든 결측치를 0으로 대체

df.fillna(0)
```

```python
# method = 'ffill' → 결측치를 이전행의 값으로 대체

df.fillna(method = 'ffill')
```

```python
# method = 'bfill' → 결측치를 다음행의 값으로 대체

df.fillna(method = 'bfill')
```

```python
# 결측치를 열마다 특정값으로 대체

values = {'A':85, 'B':80, 'C':75, 'D':90}
df.fillna(value = values)
```

```python
# 결측치를 열별 평균값으로 대체

df.fillna(value = df.mean(axis = 0))
```

```python
# 2번째 결측치까지만 결측치를 대체

df.fillna(value = values, limit = 2)
```

```python
# ★ ★ ★ 총괄 QUIZ ★ ★ ★
# (1단계) iris 각 열마다 결측치를 20개씩 인위적으로 할당한다
    # 결측치가 각 열마다 20개씩 있는지 확인
# (2-1 단계) 결측치는 각 열의 평균값으로 대체한다 : fillna 사용
# (2-2 단계) 결측치는 종별 열별 평균값(groupby, pivot_table)으로 대체한다 : fillna 사용X

import pandas as pd
import random
import seaborn as sns
import numpy as np

iris = sns.load_dataset('iris')
rowlen, collen = iris.shape
rowlen, collen
```

```python
# (1단계) iris 각 열마다 결측치를 20개씩 인위적으로 할당한다

print('임의의 20개 index 추출 :', random.sample(range(rowlen), 20))
```

```python
# 결측치 할당

for col in range(collen - 1) :
    iris.iloc[random.sample(range(rowlen), 20), col] = np.nan
```

```python
# 결측치 확인 → iris.isna(), iris.isnull() : 결측치인지 여부

# iris.isna().sum()
print(iris.isnull().sum())
iris.tail()  # 하위 5개 출력
```

```python
# 결측치 하나라도 있는 행 출력

iris[iris['sepal_length'].isna() | iris['sepal_width'].isna() |
     iris['petal_length'].isna() | iris['petal_width'].isna()]
```

```python
# (2-1단계) 결측치는 각 열의 평균값으로 대체한다 : fillna사용

# iris.mean(numeric_only = True)
# iris.iloc[:, :-1].mean()
iris.fillna(iris.iloc[:, :-1].mean())
```

```python
iris.isna().sum()
```

```python
# (2-2 단계) 결측치는 종별 열별 평균값(groupby, pivot_table)으로 대체한다 : fillna 사용X
# 종별 열별 평균값

iris_mean = iris.groupby('species').mean()
iris_mean
```

```python
# fillna 사용X 확인

iris.fillna(iris_mean)
```

```python
iris.iloc[2]
```

```python
iris_mean, iris_mean.loc['setosa', 'sepal_length']
```

```python
# iris 한행을 매개변수로 받아 결측치가 있는 경우 종별 열별 평균값으로 대체하는 함수

def iris_fillna(irisrow) :
    irisrow_copy = irisrow.copy()
    for col, value in irisrow_copy.items() :
        if isinstance(value, (int, float, np.number)) and np.isnan(value) :
            # print(irisrow[col], '이 결측치')
            irisrow[col] = iris_mean.loc[irisrow.species, col]
    return irisrow_copy
```

```python
iris_fillna(iris.iloc[2])
```

```python
# apply 적용

iris = iris.apply(iris_fillna, axis = 1)
```
